```python
import zipfile
my_zip = zipfile.ZipFile('Books.zip') # Specify your zip file's name here
storage_path = '.'
f=''
for file in my_zip.namelist():
    if my_zip.getinfo(file).filename.endswith('.txt'):
        my_zip.extract(file, storage_path) # extract the file to current folder if it is a text file
        #print(file)
        final = open(file, encoding='utf-8' )
        final = final.read().lower()
        print(type(final))
        f+=(final)
        print(len(f))
        
f=f[:100000]
print(len(f))
#print(f)


```

```python
import re
from collections import OrderedDict
f=str(f)
remove = '!()[]{};:\,<>./?@#$€%^&§*_~0123456789'
for i in f:   
    if i in remove:   
        f = f.replace(i, '')   

# Transform multiline words (with hyphen and newline) into single words
f = re.sub(r'-\n(\w+ *)', r'\1\n', f)

# Split the string f into words by space, and save them as list
f = re.split('\s', f)


# Converts the list into dictionary, where the words are the keys, and the
# number of their occurrences are their values.
def list_to_dict (words):
    words_n = {}    
    for i in words:
        x = words.count(i)
        words_n.update({i: x})
    return words_n

f = list_to_dict(f)

# Dictionary cleanup
del f['']

# Sort the dictionary by values from lowest to the highest
f = OrderedDict(sorted(f.items(), key=lambda x: x[1]))

# Output results
print (f)
```
